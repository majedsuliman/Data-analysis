```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Data1.csv")
df.head()
```

```python
dfindx = df.copy()
dfindx = dfindx.set_index('Unnamed: 0')
dfindx.head()
```

```python

f, ax = plt.subplots(3,2, figsize=(15,5))
df.plot('Unnamed: 0','theta_1', ax=ax[0][0], title = 'Theta_1')
df.plot('Unnamed: 0','theta_2', ax=ax[0][1], title = 'Theta_2')
df.plot('Unnamed: 0','theta_3', ax=ax[1][0], title = 'Theta_3')
df.plot('Unnamed: 0','theta_4', ax=ax[1][1], title = 'Theta_4')
df.plot('Unnamed: 0','theta_5', ax=ax[2][0], title = 'Theta_5')
df.plot('Unnamed: 0','theta_6', ax=ax[2][1], title = 'Theta_6')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.1, hspace=1.5)
```

```python
#Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot.
h, ax = plt.subplots(3,2, figsize=(10,15))
df.hist('theta_1', bins = 50, ax=ax[0][0])
df.hist('theta_2', bins = 50, ax=ax[0][1])
df.hist('theta_3', bins = 50, ax=ax[1][0])
df.hist('theta_4', bins = 50, ax=ax[1][1])
df.hist('theta_5', bins = 50, ax=ax[2][0])
df.hist('theta_6', bins = 50, ax=ax[2][1])
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
```

```python
df.plot.kde()
```

```python
df['Unnamed: 0'] = pd.to_datetime(df['Unnamed: 0'], format='%Y-%m-%d')
Year18_df = df.loc[(df['Unnamed: 0'] >= '2018-01-01') & (df['Unnamed: 0'] <= '2018-12-31')]
theta_4 = Year18_df[['Unnamed: 0', 'theta_1', 'theta_2','theta_3', 'theta_4']]
theta_4.set_index('Unnamed: 0')
```

```python
#Plot all columns as time series. (Year 2018)
f1, ax = plt.subplots(2,2, figsize=(15,5))
theta_4.plot('Unnamed: 0','theta_1', ax=ax[0][0], title = 'Theta_1')
theta_4.plot('Unnamed: 0','theta_2', ax=ax[0][1], title = 'Theta_2')
theta_4.plot('Unnamed: 0','theta_3', ax=ax[1][0], title = 'Theta_3')
theta_4.plot('Unnamed: 0','theta_4', ax=ax[1][1], title = 'Theta_4')
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.1, hspace=1.5)
```

```python

#Plot histograms of all columns, verify bin size. Plot all on a single, faceted plot. (Year 2018)
h1, ax = plt.subplots(2,2, figsize=(10,12))
theta_4.hist('theta_1', bins = 20, ax=ax[0][0])
theta_4.hist('theta_2', bins = 20, ax=ax[0][1])
theta_4.hist('theta_3', bins = 20, ax=ax[1][0])
theta_4.hist('theta_4', bins = 20, ax=ax[1][1])
plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.3)
```

```python
theta_4.set_index('Unnamed: 0', inplace = True)
theta_4.plot.kde()
```
