```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("Data1.csv")
df.head()
```

```python
dfindx = df.copy()
dfindx = dfindx.set_index('Unnamed: 0')
dfindx.head()
```

```python
dfindx.plot( y=["theta_1"], use_index = True)
dfindx.plot( y=["theta_2"], use_index = True)
dfindx.plot( y=["theta_3"], use_index = True)
dfindx.plot( y=["theta_4"], use_index = True)
dfindx.plot( y=["theta_1"], use_index = True)
plt.show()
```

```python
binNum = 3
x= [df['theta_1'],df['theta_2'],df['theta_3'],df['theta_4'],df['theta_5'],df['theta_6']]
plt.hist(x, bins = binNum)
plt.legend(["theta_1","theta_2","theta_3","theta_4","theta_5","theta_6"])
plt.show()
```
